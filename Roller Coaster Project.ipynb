Write a function that will plot the ranking of the top n ranked roller coasters over time as lines. Your function should take a number n and a ranking DataFrame as arguments. Make sure to include informative labels that describe your visualization.

For example, if n == 5, your function should plot a line for each roller coaster that has a rank of 5 or lower.

```python
plt.clf()

def top_rank(df, n):
    top_df = df[df['Rank'] <= n]
    fig, ax = plt.subplots(figsize=(20,10))
    for coaster in set(top_df['Name']):
        rank = top_df[top_df['Name']==coaster]
        plt.plot(rank['Year of Rank'], rank['Rank'], label=coaster, marker='o')
    ax.set_yticks([i for i in range(1,n)])
    ax.invert_yaxis()
    plt.title('Roller Coasters with a Rank of {} or Higher'.format(n))
    plt.xlabel('Year')
    plt.ylabel('Ranking')
    plt.legend()
    plt.show()

top_rank(steel, 5)
```

Write a function that plots a histogram of any numeric column of the roller coaster DataFrame. Your function should take a DataFrame and a column name for which a histogram should be constructed as arguments. Make sure to include informative labels that describe your visualization.

```python
def histogram(df, df_c):
    plt.hist(df[df_c].dropna())
    plt.title('{} Histogram'.format(df_c).upper())
    plt.xlabel(df_c)
    plt.ylabel('Number of Roller Coasters')
    plt.show()

###Test
histogram(rcdata, 'speed')
```

```python
###Create a function to plot histogram of height values
def height_histogram(df):
    coaster = df[df['height']<=180]['height'].dropna()
    plt.hist(coaster, bins=20)
    plt.title('Histogram of Coaster Height')
    plt.xlabel('Height')
    plt.ylabel('Number of Roller Coasters')
    plt.show()

### Plot it
height_histogram(rcdata)
```

Write a function that creates a bar chart showing the number of inversions for each roller coaster at an amusement park. Your function should take the roller coaster DataFrame and an amusement park name as arguments. Make sure to include informative labels that describe your visualization.

```python
def bar_inv(df, parkn):
    park_coasters = df[df['park'] == parkn]
    names = park_coasters['name']
    num_inversions = park_coasters['num_inversions']
    fig, ax = plt.subplots(figsize=(20,10))
    plt.bar(range(len(park_coasters)), num_inversions)
    ax.set_xticks(range(len(park_coasters)))
    ax.set_xticklabels(names, rotation = 45)
    plt.title('Number of Inversions in {}'.format(parkn))
    plt.xlabel('Roller Coaster')
    plt.ylabel('Num of Inversions')
    plt.show()

bar_inv(rcdata, 'Six Flags Great Adventure')
```

Write a function that creates a pie chart that compares the number of operating roller coasters ('status.operating') to the number of closed roller coasters ('status.closed.definitely'). Your function should take the roller coaster DataFrame as an argument. Make sure to include informative labels that describe your visualization.

```python
def pie_ch(df):
    operating = len(df[df['status'] == 'status.operating'])
    closed = len(df[df['status'] == 'status.closed.definitely'])
    all_coasters = [operating, closed]
    plt.pie(all_coasters, autopct = '%0.f%%', labels = ['Operating', 'Closed'])
    plt.axis('equal')
    plt.title('Percentage of Operating and Closed Roller Coasters')
    plt.show()

pie_ch(rcdata)
```

Create a function to plot scatter of speed vs height



```python
def scatter_plot(df, col1, col2):
    fig, ax5 = plt.subplots(figsize=(20,10))
    plt.scatter(df[col1], df[col2])
    plt.title('{} vs {}'.format(col1, col2))
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()

scatter_plot(rcdata, 'speed', 'height')
```
