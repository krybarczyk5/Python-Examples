```python
### Needed libraries
import seaborn
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Pulling in the webpage and using an html parser to read it
webpage = requests.get('https://content.codecademy.com/courses/beautifulsoup/cacao/index.html')
soup = BeautifulSoup(webpage.content, "html.parser")
print(soup)
```

```python
### Getting all of the ratings from the webpage
rating_tags = soup.find_all(attrs={"class": "Rating"})
print(rating_tags)
```

```python
### Taking the data and putting it into a list
ratings = []
for text in rating_tags[1:]:
  ratings.append(float(text.get_text()))

print(ratings)
```

```python
### Making a histogram to view the rating data
plt.hist(ratings)
plt.show()
```

```python
### Getting all of the company names form the table
company_tags = soup.find_all(attrs={"class": "Company"})
print(company_tags)
```

```python
### Putting the company data into a list
companies = []
for text in company_tags[1:]:
  companies.append(text.get_text())
print(companies)
```

```python
### Combining the companies and their associated ratings into a dataframe for analysis
df = {"Company": companies, "Rating": ratings}
choc_df = pd.DataFrame.from_dict(df)
print(choc_df)
```

```python
### Getting the average rating of chocolate by company and further breaking it down to show the top ten
avg_rating = choc_df.groupby("Company").Rating.mean()
top_ten = avg_rating.nlargest(10)
print(avg_rating)
print(top_ten)
```

```python
### Getting the cocoa percentage values formthe table
cocoap_tags = soup.find_all(attrs={"class": "CocoaPercent"})
print(cocoap_tags)
```

```python
### Putting the cocoa percentages into a list and removing the % sign
cocoap = []
for text in cocoap_tags[1:]:
  cocoap.append(float(text.get_text().strip('%')))
print(cocoap)
```

```python
### Adding our percentage data to our dataframe
df['CocoaPercentage'] = cocoap
choc_df = pd.DataFrame.from_dict(df)
print(choc_df)
```

```python
### Creating a scatter plot with a trend line to see if there is any association
### between the percentage of cocoa and the rating the chocolate receives
plt.clf()
plt.scatter(choc_df.CocoaPercentage, choc_df.Rating)
z = np.polyfit(choc_df.CocoaPercentage, choc_df.Rating, 1)
line_function = np.poly1d(z)
plt.plot(choc_df.CocoaPercentage, line_function(choc_df.CocoaPercentage), "r--")
plt.show()
```
